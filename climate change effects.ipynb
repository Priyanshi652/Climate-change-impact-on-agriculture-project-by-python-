"AgroEarth" is a leading agricultural advisory firm that has noticed a decline in crop yields across various regions over the past decade.

AgroEarth aims to identify the underlying causes and provide actionable solutions to enhance yield stability, optimize resource management, and promote sustainable farming practices in the face of climate change.

IMPORTING NECESSARY LIBRARIES and LOADING THE DATASET-:-:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r"C:\Users\DELL\Desktop\climate_change_impact_on_agriculture_2024.csv")
df.head(5)
```

CHECKING FOR MISSING VALUE

```python
df.drop(labels="Extreme_Weather_Events",axis=1,inplace=True)
```

```python
df.info()
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

CHECKING THE DATATYPE

```python
df.dtypes 
```

CONVERTING VALUES ACCORDING TO THE DATATYPE(if not)

```python
df['CO2_Emissions_MT'] = df['CO2_Emissions_MT'].astype('float')

df['Region'] = df['Region'].astype('category')

df['Average_Temperature_C'] = df['Average_Temperature_C'].astype('float')

df['Crop_Yield_MT_per_HA'] = df['Crop_Yield_MT_per_HA'].astype('float')

df['Crop_Type'] = df['Crop_Type'].astype('category')

df['Year'] = pd.to_datetime(df['Year'], format='%Y')

df['Total_Precipitation_mm '] = df['Total_Precipitation_mm'].astype('float')

df['Soil_Health_Index'] = df['Soil_Health_Index'].astype('category') 
```

1. Yearly Impact On Economics-:

we will use lineplot to show the trend of impact.

```python
plt.figure(figsize=(28, 8))
sns.lineplot(data=df, x='Year', y='Economic_Impact_Million_USD',marker='*',markersize=15,color="brown")
plt.title('Yearly Impact On Economy')
plt.xlabel('Year')
plt.ylabel('Economic impact (im million usd)')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Irrigation_Access_%', y='Crop_Yield_MT_per_HA', hue='Country', alpha=0.9, palette='viridis')

plt.title('Relationship Between Irrigation Access and Crop Yields by Country')
plt.xlabel('Irrigation Access')
plt.ylabel('Crop Yield (tons/hectare)')
plt.grid(True)

plt.legend(title='Country')
plt.show()
```

3. Identifying the Impact of Climate Variables on Crop Yields to Optimize Agricultural Practices.

   
We will a heatmap to show the correlation between temperature, rainfall, CO2, and crop yields.

```python
correlation=df[["Average_Temperature_C","Crop_Yield_MT_per_HA","CO2_Emissions_MT","Total_Precipitation_mm"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='YlGn', linewidths=0.5)
plt.title('Correlation Between Climate Variables and Crop Yields')
plt.show()
```

4.Analyzing the Impact of CO2 Emissions on Crop Yield Across Different Countries Over the Past 5 Years 
we will use  bar charts to show the crops and regions with the largest decline in yield


```python
region_adaptation_yield = df.groupby(['Year','Country'])['CO2_Emissions_MT'].sum().unstack()
sort=region_adaptation_yield.sort_values(by='Year').tail(5)
sort.plot(kind='bar', stacked=True, figsize=(10, 6), colormap="Greens")
plt.title('CO2 Emmision in different countries of Past 5 years')
plt.xlabel('Years')
plt.ylabel('CO2_Emissions_MT')
plt.legend(loc='best', bbox_to_anchor=(1,1), title='COUNTRIES')
plt.show()
```

5. Assessing the Economic Impact of Various Climate Adaptation Strategies to Inform Sustainable Investment Decisions.

we will use pie chart for the statement.

```python
strategy_data = df.groupby('Adaptation_Strategies')['Economic_Impact_Million_USD'].sum().reset_index()

# Extracting labels and sizes from the grouped data
labels = strategy_data['Adaptation_Strategies']
sizes = strategy_data['Economic_Impact_Million_USD']

# Customizing the pie chart
colors = ['#556B2F', '#6B8E23', '#8B4513', 'lightgreen', '#228B22']
explode = (0, 0.1, 0.1, 0, 0)  # Exploding the largest slice for emphasis

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

# Equal aspect ratio ensures the pie chart is circular
plt.axis('equal')
plt.title('Econmic Impact by Different Adaption Strategies')
plt.show()
```

6. Identifying Yield Instability Across Regions to Enhance Agricultural Risk Management and Stability

we will use horizontal bar chart to find the 5 most unstable crops.

```python
# Calculate the standard deviation of crop yields for each region
region_stability = df.groupby(['Crop_Type','Region'])['Crop_Yield_MT_per_HA'].std()

# Sort regions by instability (higher standard deviation means more unstable)
unstable_regions = region_stability.sort_values(ascending=False).head(5)

# Plot the top 5 unstable regions
plt.figure(figsize=(10, 6))
unstable_regions.plot(kind='barh', color='darkgreen')
plt.title('Top 5 Most Unstable Regions (Higher = More Unstable)')
plt.ylabel('Region')
plt.xlabel('Yield Standard Deviation (tons/hectare)')
plt.show()

# Display results for the top 5 unstable regions
print("Top 5 regions with the most unstable crop yields:")
for region, value in unstable_regions.items():
    print(f"Region: {region}, Standard Deviation: {value:.2f} tons/hectare")
```

7. Analyzing Precipitation Patterns Across Crop Types to Enhance Crop Resilience and Yield Optimization.

we will use boxplot for the statement.

```python
# Box plot for Precipitation vs Crop Yield by Crop Type
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Crop_Type', y='Total_Precipitation_mm', palette='Set2')
plt.title('Precipitation vs Crop Types')
plt.xlabel('Crop Type')
plt.ylabel('Total Precipitation (mm)')
plt.xticks(rotation=45)
plt.show()
```

8. Evaluating Pesticide and Fertilizer Consumption Across Crop Types to Enhance Resource Efficiency and Environmental Sustainability

we will use bar chart for the fertilizers and pesticides usage

```python
# Group the data by 'Crop_Type' and get the mean of Pesticide and Fertilizer usage
grouped_data = df.groupby('Crop_Type')[['Pesticide_Use_KG_per_HA', 'Fertilizer_Use_KG_per_HA']].mean()
colors=["brown","green"]

# Create a grouped bar plot directly using pandas' plot
grouped_data.plot(kind='bar', figsize=(10,6), width=0.8,color=colors)

# Add labels and title
plt.title('Pesticide and Fertilizer Usage by Crop Type')
plt.xlabel('Crop Type')
plt.ylabel('Usage (KG/HA)')
plt.xticks(rotation=45)
plt.legend(['Pesticide Usage (KG/HA)', 'Fertilizer Usage (KG/HA)'])

# Show plot
plt.tight_layout()
plt.show()
```

9. Assessing the Relationship Between Average Temperature and CO2 Emissions by Region to Inform Climate Mitigation Strategies

we will use subplot(linechart) for identifying the relation.

```python
import matplotlib.pyplot as plt

# Group the data by 'Region' and calculate the mean of Irrigation Access and CO2 Emissions
grouped_data = df.groupby('Region')[['Average_Temperature_C', 'CO2_Emissions_MT']].mean()

# Create the figure and axis objects
fig, ax1 = plt.subplots(figsize=(40, 6)) 

# Plotting Irrigation Access on the first y-axis (left)
ax1.plot(grouped_data.index, grouped_data['Average_Temperature_C'], color='green',markersize=10, marker='o', label='Average Temperature',linewidth=2)
ax1.set_xlabel('Region')
ax1.set_ylabel('verage_Temperature_C', color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Plotting CO2 Emissions on the second y-axis (right)
ax2 = ax1.twinx()
ax2.plot(grouped_data.index, grouped_data['CO2_Emissions_MT'], color='brown',markersize=10,marker='s', label='CO2 Emissions (MT)',linewidth=2)
ax2.set_ylabel('CO2 Emissions (MT)', color='brown')
ax2.tick_params(axis='y', labelcolor='brown')

# Adjust the rotation and font size of x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=8)

# Add title and ensure layout is tight
plt.title('Comparison of Average Temperature and CO2 Emissions by Region')
plt.tight_layout()
plt.grid(True)
# Show plot
plt.show()
```
